```python
# Importar bibliotecas pandas e numpy #

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extrair base de dados e mostra-los #

dataframe = pd.read_csv("transito.csv")
dataframe.head()
```

```python
# Informações da tabela#

dataframe.info()
```

```python
# Checando valores nulos#

ValoresNulos = dataframe.isna().sum()
print (ValoresNulos)
```

```python
# Excluir dados duplicados

dataframe_sem_duplicatas = dataframe.drop_duplicates(subset=['origem'])
```

```python
dataframe_sem_duplicatas.head()
```

```python
# Criar um gráfico de dispersão

plt.figure(figsize=(8, 6))
plt.scatter(dataframe['distancia'], dataframe['duracao'], color='blue', alpha=0.5)

# Adicionar rótulos e título

plt.title('Duração x Distância')
plt.xlabel('Distância (km)')
plt.ylabel('Duração (min)')
plt.grid(True)

# Exibir o gráfico

plt.show()
```

```python
# Criar o gráfico de linha

plt.figure(figsize=(8, 6))
plt.plot(dataframe['distancia'], dataframe['duracao'], marker='o', color='blue', linestyle='-')

# Adicionar rótulos e título

plt.title('Duração em relação à Distância')
plt.xlabel('Distância (km)')
plt.ylabel('Duração (min)')
plt.grid(True)

# Exibir o gráfico

plt.show()
```
